```python
import folium


m = folium.Map(location=[47.800965, 13.035831])
m
```

```python
m.save("index.html")
```

```python
folium.Map(location=[47.800965, 13.035831], zoom_start=20)
```

```python
m = folium.Map(location=[47.800965, 13.035831], zoom_start=12, tiles="OpenStreetMap")

tooltip = "Click me!"

folium.Marker(
    [47.82122849185476, 13.037804923371962], popup="<i>Asvö Sportplatz Itzling</i>", tooltip=tooltip
).add_to(m)
folium.Marker(
    [47.82104744799504, 13.054023401429816], popup="<b>PADELBASE Salzburg</b>", tooltip=tooltip
).add_to(m)

m
```
